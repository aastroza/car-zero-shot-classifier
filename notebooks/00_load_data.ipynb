```python
import pandas as pd
from cappr.openai.classify import predict
```

```python
df = pd.read_csv('../data/raw/nodes_marca_modelo.csv')
```

```python
df.head()
```

```python
#new column with concatenation of marca and modelo
df['marca_modelo'] = df['marca'] + ' ' + df['modelo']
```

```python
# get first value of marca_modelo
marca_modelo = df['marca_modelo'].iloc[0]
marca_modelo
```

```python
# get list of values of marca_modelo
marca_modelos = df['marca_modelo'].tolist()
```

```python
car_types = ['Sedan', 'SUV', 'Convertible', 'Crew Cab', 'Van', 'Hatchback', 'Coupe', 'Station Wagon']
```

```python
translation_dict = {
    'Sedan': 'Sedán',
    'SUV': 'SUV',
    'Convertible': 'Convertible',
    'Crew Cab': 'Doble Cabina',
    'Van': 'Furgón',
    'Hatchback': 'Hatchback',
    'Coupe': 'Coupé',
    'Station Wagon': 'Station Wagon'
}
```

```python
result = []

for car in marca_modelos:
    prompt = f'''As a car expert, which of the following categories would you most likely classify a {car} into?\n Car Type:'''
    class_names = ('Sedan', 'SUV', 'Convertible', 'Crew Cab', 'Van', 'Hatchback', 'Coupe', 'Station Wagon')
    # optional: let's supply a prior distribution over the classes
    #prior       = (   1/8    ,    1/8   ,     3/4   )

    preds = predict(prompts=[prompt],
                    completions=class_names,
                    model='text-davinci-003')

    preds = predict(prompts=[prompt],
                    completions=class_names,
                    model='text-davinci-003')#second time because of cache bug
    result.append(preds)
```

```python
df['tipo_gpt'] = [r[0] for r in result]
```
